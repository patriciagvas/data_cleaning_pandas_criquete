```python
import pandas as pd
from google.colab import drive
```

```python
# Importar base de dados

df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Datasets/Base de dados projeto cricket.xlsx')
df.head(5)
```

```python
df.shape
```

```python
# Renomear colunas para melhor compreensão dos dados
df= df.rename(columns = {'NO': 'Not_Out', 'HS': 'Highets_Inns_Score', 'BF': 'Balls_Faced', 'SR': 'Batting_Strike_Rate', 'Mat': 'Matches'})
df.head(10)
```

```python
# Verificar valores nulos

df.isnull().any()
```

```python
# Substituir - por 0 na coluna Balls_Faced
df['Balls_Faced'] = df['Balls_Faced'].replace('-', 0)
print(df.Balls_Faced)
```

```python
# Eliminar duplicatas
df.duplicated()
```

```python
# Identificar linhas duplicadas
df[df['Player'].duplicated()==1]
```

```python
# Eliminar duplicatas
df= df.drop_duplicates()
```

```python
# Separar Span em data de início e data final
df['Span'].str.split(pat = '-')
```

```python
df['Rokie_Year'] = df['Span'].str.split(pat = '-').str[0]
```

```python
df['Final_Season'] = df['Span'].str.split(pat = '-').str[1]
df.head()
```

```python
# Eliminar coluna Span
df = df.drop(['Span'], axis = 1)
```

```python
df.head()
```

```python
# Separar informçãoes dos jogadores em uma coluna com seus nomes e seu país de origem
df['Country'] = df['Player'].str.split(pat = '(').str[1]
df['Country'] = df['Country'].str.split(pat = ')').str[0]
df.head()
```

```python
df['Player'] = df['Player'].str.split(pat = '(').str[0]
df.head()
```

```python
# Alterar o tipo dos dados
df.dtypes
```

```python
df['Highets_Inns_Score'] = df['Highets_Inns_Score'].apply(lambda x: x.replace('*', '') if isinstance(x, str) else x)
print(df.Highets_Inns_Score)
```

```python
df['Highets_Inns_Score']= pd.to_numeric(df['Highets_Inns_Score'], errors='coerce')
```

```python
df.dtypes
```

```python
df['Balls_Faced'] = df['Balls_Faced'].apply(lambda x: x.replace('+', '') if isinstance(x, str) else x)
print(df.Balls_Faced)
```

```python
df['Balls_Faced'] = df['Balls_Faced'].astype('int')
df.dtypes
```

```python
df = df.astype({'Rokie_Year': 'int', 'Final_Season': 'int'})
df.dtypes
```

```python
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].astype('float')
df.dtypes
```

```python
# Adicionar uma coluna com o total de anos de carreira no esporte
df['Career_Lenght'] = df['Final_Season'] - df['Rokie_Year']
print(df.Career_Lenght)
```

```python
# Média de anos de carreira
media = round(df['Career_Lenght'].mean(),2)
print(f"A média de anos de carreira dos jogadores é {media} anos.")
```

```python
# Batting_Strike_Rate de jogadores com mais de 10 anos de carreira
media_batting = round(df[df['Career_Lenght'] > 10]['Batting_Strike_Rate'].mean(), 2)
print(f"A média de Batting Strike Rate para jogadores com mais de 10 anos de carreira é {media_batting}")
```

```python
# Total de jogadores que jogaram antes de 1960
total_jogadores = df[df['Rokie_Year'] < 1960]['Player'].count()
print(f"O total de jogadores que jogaram antes de 1960 é {total_jogadores}")
```

```python
# Maior Highest Inns Score por país
inns_country = df.groupby('Country')['Highets_Inns_Score'].max().to_frame('Highinncountry').reset_index().sort_values('Highinncountry', ascending = False)
print(inns_country)
```
